```python
import pandas as pd
nba_file = "https://raw.githubusercontent.com/EduPekUfpr/MNUM-7124/main/Data/nba.csv"
revenue_file = "https://raw.githubusercontent.com/EduPekUfpr/MNUM-7124/main/Data/revenue.csv"

nba = pd.read_csv(nba_file)
nba.head()
```

## Value counts

```python
nba["Position"].value_counts()
```

```python
nba["Salary"].value_counts()
```

## Missing Values

```python
nba = pd.read_csv(nba_file)
nba.tail()
```

```python
nba.dropna() #deleta todas as linhas que contem pelo menos um NaN
```

```python
nba.dropna(how = "any") #deleta todas as linhas que contem pelo menos um NaN
```

```python
nba.dropna(how = "all") #deleta todas as linhas que contem TODOS os valores NaN
```

```python
nba.dropna(subset = ["College"]) #deleta todas as linhas que contem NaN em colunas específicas
```

```python
nba.dropna(subset = ["College","Salary"]) #deleta todas as linhas que contem NaN em colunas específicas
```

# Fill In

```python
nba = pd.read_csv(nba_file)
nba.tail()
```

```python
nba.fillna(0)
```

```python
nba["College"].fillna("UFPR")
```

```python
nba["College"].fillna("UFPR", inplace = True)
nba
```

```python
nba["Salary"].fillna(value = nba["Salary"].mean(), inplace = True)
nba
```

## As Type

```python
nba = pd.read_csv(nba_file).dropna(how = "all")
nba.head()
```

```python
nba["Age"].hasnans
```

```python
nba["Age"].astype("int")
```

```python
nba["Age"] = nba["Age"].astype(int)
nba
```

```python
nba["Salary"].hasnans
```

```python
nba["Salary"].astype(int)
```

```python
nba["Salary"].fillna(value = nba["Salary"].mean(), inplace = True)
nba["Salary"].astype(int)
nba
```

## Categorical Data

```python
nba = pd.read_csv(nba_file).dropna(how = "all")
nba.head()
```

```python
nba.dtypes
```

```python
nba["Position"].unique()
```

```python
nba["Position"].nunique()
```

```python
nba.info()
```

```python
nba["Position"] = nba["Position"].astype("category")
nba.info()
```

```python
nba["Team"] = nba["Team"].astype("category")
nba.info()
```
