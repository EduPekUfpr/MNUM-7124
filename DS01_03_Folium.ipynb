````python
import pandas as pd
import folium
````

````python
latO = city.loc[0].at["city_lat"]
lonO = city.loc[0].at["city_lon"]
nameO = city.loc[0].at["city"]

print(nameO)

m = folium.Map(location=[city.loc[0].at["city_lat"], city.loc[0].at["city_lon"]])

#plot Origin
folium.Marker(
    location=[latO, lonO],
    popup=nameO,
    icon=folium.Icon(color="blue"),
            ).add_to(m)

# Just to zoom the maps' bounds
m.fit_bounds(m.get_bounds())

#plot map
m
````

## Choropleth 

Usa o polígono e preenche com algum dado

````python
url = ("https://raw.githubusercontent.com/python-visualization/folium/main/examples/data")

state_geo = f"{url}/us-states.json"
state_unemployment = f"{url}/US_Unemployment_Oct2012.csv"
state_data = pd.read_csv(state_unemployment)

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=state_geo,
    name="choropleth",
    data=state_data,
    columns=["State", "Unemployment"],
    key_on="feature.id",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Unemployment Rate (%)",
).add_to(m)

folium.LayerControl().add_to(m)

m
````

## Rotas 

* a biblioteca ```Nominatim``` encontra as coordenadas com o nome da cidade e vice-e-versa.
* O método ```osmnx.graph_from_point``` faz o dowload de todos os pontos em torno de uma coordenada e tranforma nu grafo. A distância é dada pelo atributo ```dist``` em metros. Vocês devem imaginar que se este atributo for muito grande a teremos muitos pontos, mas muitos mesmos.
* O método ```networkx.shortest_path``` aplica o algoritmo de Djikstra (se pronuncia Dáistra) ao grafo gerado acima. então se tivermos muitos pontos a resolução irá demorar muito.
  

````python
import osmnx as ox
import networkx as nx
from geopy.geocoders import Nominatim

# Initialize the geolocator
geolocator = Nominatim(user_agent="route_map")

# Get the coordinates for São José dos Pinhais
cwb_location = geolocator.geocode("Curitiba, PR, Brazil")
cwb_coords = (cwb_location.latitude, cwb_location.longitude)

# Get the coordinates for Curitiba
sjp_location = geolocator.geocode("São José dos Pinhais, PR, Brazil")
sjp_coords = (sjp_location.latitude, sjp_location.longitude)

# Download the road network graph around Curitiba
G = ox.graph_from_point(cwb_coords, dist=12000, network_type='drive')

# Get the nearest nodes to the start and end points
orig_node = ox.distance.nearest_nodes(G, cwb_coords[1], cwb_coords[0])
dest_node = ox.distance.nearest_nodes(G, sjp_coords[1], sjp_coords[0])

# Find the shortest path using Dijkstra's algorithm
route = nx.shortest_path(G, orig_node, dest_node, weight='length')

# Extract the latitude and longitude for the nodes in the route
route_points = [(G.nodes[node]['y'], G.nodes[node]['x']) for node in route]

# Create a Folium map centered at Curitiba
route_map = folium.Map(location=cwb_coords)

# Add the route to the map
folium.PolyLine(route_points, color='blue', weight=2.5, opacity=1).add_to(route_map)

# Add markers for New York City and Buffalo
folium.Marker(cwb_coords, popup='Curitiba').add_to(route_map)
folium.Marker(sjp_coords, popup='São José dos Pinhais').add_to(route_map)

# Correct the zoom
route_map.fit_bounds(route_map.get_bounds())

# Save the map to an HTML file
route_map.save("cwb_sjp_route.html")

# Show the map
route_map
````
